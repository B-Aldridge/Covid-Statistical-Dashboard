```python
from splinter import Browser
from selenium.webdriver.chrome.service import Service
driver = Service(executable_path='C:\chromedriver_win32\chromedriver.exe')
from bs4 import BeautifulSoup as soup
import pandas as pd
from pprint import pprint
from urllib.error import HTTPError

browser = Browser('chrome', service=driver)

# Visit WHO website
url = 'https://covid19.who.int/data'
browser.visit(url)

# Get the page HTML content
html = browser.html
sp = soup(html, "html.parser")

# Find all link elements
link_elements = sp.find_all('a')

# Extract csv links that hold relevant data
# (There is one other csv link, but it doesn't end in "data")
csv_links=[]
links=[]
for link in link_elements:
   links.append(link['href'])
   if 'data.csv' in link['href']:
        csv_links.append(link['href'])

pprint(f'FROM: {links} THE FOLLOWING CSV LINKS WERE SELECTED: {csv_links}')


# For each csv link, read that csv into pandas, extract its name from the URL, and save as csv to repository
for link in csv_links:
    try: csv=pd.read_csv(link,index_col=False)
    except HTTPError:
        print(f"LINK {link} DOES NOT WORK!")
    else:
        name=link[(link.rfind('/')+1):]
        csv.to_csv(f'CSV/{name}')
        print(f'{name} IMPORTED SUCCESSFULLY!')
```
