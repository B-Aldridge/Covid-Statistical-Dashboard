```python
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup as soup

browser = Browser('chrome', executable_path=ChromeDriverManager().install())

# Visit worldmeters for data
url = 'https://www.worldometers.info/coronavirus/'
browser.visit(url)

# Get the page HTML content
html = browser.html
sp = soup(html, "html.parser")

# Find the table element by its ID
table = sp.find('table', {'id': 'main_table_countries_today'})

# Extract total cases
total_cases = table.find('tr', class_='total_row_world').find('td', class_='sorting_1').text.strip()

print("Total Cases:", total_cases)
```

```python
# Extract the column headers
headers = table.find('thead').find_all('th')
column_headers = [header.text.strip() for header in headers[2:]]  

# Extract the data from the table
rows = table.find('tbody').find_all('tr')
data = []
for row in rows:
    cells = row.find_all('td')
    row_data = [cell.text.strip() for cell in cells[1:]]  
    data.append(row_data)

# Extract the total cases, new cases, total deaths, etc. from the data
total_cases = [row[0] for row in data]
new_cases = [row[1] for row in data]
total_deaths = [row[2] for row in data]
new_deaths = [row[3] for row in data]
total_recovered = [row[4] for row in data]
new_recovered = [row[5] for row in data]
active_cases = [row[6] for row in data]
serious_critical = [row[7] for row in data]

# Print the extracted cases
print("Total Cases:", total_cases)
print("New Cases:", new_cases)
print("Total Deaths:", total_deaths)
print("New Deaths:", new_deaths)
print("Total Recovered:", total_recovered)
print("New Recovered:", new_recovered)
print("Active Cases:", active_cases)
print("Serious/Critical:", serious_critical)

```

```python
import pandas as pd
# Create a DataFrame
data = {
    'Total Cases': total_cases,
    'New Cases': new_cases,
    'Total Deaths': total_deaths,
    'New Deaths': new_deaths,
}

df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```
